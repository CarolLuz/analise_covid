```python
import pandas as pd
import seaborn as sns
sns.set_theme()
import matplotlib.pylab as plb
```

# 1. BASE DE DADOS

```python
covid2020_1 = pd.read_csv('covid_2020_1.csv')
covid2020_2 = pd.read_csv('covid_2020_2.csv')
```

```python
covid2021_1 = pd.read_csv('covid_2021_1.csv')
covid2021_2 = pd.read_csv('covid_2021_2.csv')
```

```python
covid2022_1 = pd.read_csv('covid_2022_1.csv')
covid2022_2 = pd.read_csv('covid_2022_2.csv')
```

```python
covid2020_mensal = pd.read_csv('covid_2020_mensal.csv')
```

```python
covid2021_mensal = pd.read_csv('covid_2021_mensal.csv')
```

```python
covid2022_mensal = pd.read_csv('covid_2022_mensal.csv')
```

# 2. UNINDO TABELAS

```python
covid_semestral = pd.concat([
    covid2020_1, 
    covid2020_2, 
    covid2021_1, 
    covid2021_2, 
    covid2022_1, 
    covid2022_2], axis=0).sort_values(by='data').reset_index(drop=True)

covid_semestral = covid_semestral.drop(columns=['municipio', 'regiao'])

covid_semestral.head()
```

```python
covid_mensal = pd.concat([
    covid2020_mensal,
    covid2021_mensal,
    covid2022_mensal
], axis=0).sort_values(by='data').reset_index(drop=True)

covid_mensal
```

# 3. ANÁLISE DE DADOS

```python
grafico_casos = sns.lineplot(x="data", 
y="casosAcumulado", 
data=covid_semestral, 
hue='estado').set(title='Casos acumulados por semestre - 2020 a 2021')

grafico_casos= plb.xticks(rotation=45)
```

```python
grafico_obitos = sns.lineplot(x="data", 
y="obitosAcumulado", 
data=covid_semestral, 
hue='estado').set(title='Óbitos acumulados por semestre - 2020 a 2021')

grafico_obitos= plb.xticks(rotation=45)
```

```python
grafico_casos_mes = sns.lineplot(x="data", 
y="casosAcumulado", 
data=covid_mensal, 
hue='estado').set(title='Casos acumulados por mês - 2020 a 2022')

grafico_casos_mes= plb.xticks(rotation=45)
```

```python
grafico_obitos_mes = sns.lineplot(x="data", 
y="obitosAcumulado", 
data=covid_mensal, 
hue='estado').set(title='Óbitos acumulados por mês - 2020 a 2022')

grafico_obitos_mes= plb.xticks(rotation=45)
```
