### 2.3. 2022.1: dados do último dia do semestre (30 de junho)

### 2.5. 2022.2 : filtro para dados do nordeste

### 2.6. 2022.2: filtro para dados estaduais

```python
b2022_2 = b2022_2[b2022_2['municipio'].isna()]
b2022_2
```

### 2.7. 2022.2: dados do último dia do semestre (07 de novembro)

```python
b2022_2_ultimo_dia = b2022_2.loc[(b2022_2['data']=='2022-11-07')]
b2022_2_ultimo_dia
```

```python
b2022_2_ultimo_dia = b2022_2_ultimo_dia[b2022_2_ultimo_dia['codmun'].isna()]
b2022_2_ultimo_dia
```

```python
b2022_2_ultimo_dia = b2022_2_ultimo_dia.drop(columns=['coduf', 'codmun', 'codRegiaoSaude', 'nomeRegiaoSaude', 'Recuperadosnovos', 'emAcompanhamentoNovos', 'interior/metropolitana'])
b2022_2_ultimo_dia
```

### 2.9. 2022: informações mensais (primeiro dia dos meses)

```python
jan = b2022_1.loc[b2022_1['data']=='2022-01-01']
jan = jan.loc[jan['codmun'].isna()].sort_values(by='estado')

fev = b2022_1.loc[b2022_1['data']=='2022-02-01']
fev = fev.loc[fev['codmun'].isna()].sort_values(by='estado')

mar = b2022_1.loc[b2022_1['data']=='2022-03-01']
mar = mar.loc[mar['codmun'].isna()].sort_values(by='estado')

abr = b2022_1.loc[b2022_1['data']=='2022-04-01']
abr = abr.loc[abr['codmun'].isna()].sort_values(by='estado')

mai = b2022_1.loc[b2022_1['data']=='2022-05-01']
mai = mai.loc[mai['codmun'].isna()].sort_values(by='estado')

jun = b2022_1.loc[b2022_1['data']=='2022-06-01']
jun = jun.loc[jun['codmun'].isna()].sort_values(by='estado')

jul = b2022_2.loc[b2022_2['data']=='2022-07-01']
jul = jul.loc[jul['codmun'].isna()].sort_values(by='estado')

ago = b2022_2.loc[b2022_2['data']=='2022-08-01']
ago = ago.loc[ago['codmun'].isna()].sort_values(by='estado')

setb = b2022_2.loc[b2022_2['data']=='2022-09-01']
setb = setb.loc[setb['codmun'].isna()].sort_values(by='estado')

out = b2022_2.loc[b2022_2['data']=='2022-10-01']
out = out.loc[out['codmun'].isna()].sort_values(by='estado')

nov = b2022_2.loc[b2022_2['data']=='2022-11-01']
nov = nov.loc[nov['codmun'].isna()].sort_values(by='estado')

b2022_meses = pd.concat([
    jan, 
    fev, 
    mar, 
    abr, 
    mai, 
    jun, 
    jul, 
    ago, 
    setb, 
    out, 
    nov], axis=0)

b2022_meses = b2022_meses.sort_values('data')

b2022_meses
```

### 2.10. 2022: informações mensais - removendo colunas não informativas

```python
b2022_meses = b2022_meses.drop(columns=[
    'semanaEpi',
    'regiao',
    'municipio', 
    'populacaoTCU2019',
    'coduf', 
    'codmun', 
    'codRegiaoSaude', 
    'nomeRegiaoSaude', 
    'Recuperadosnovos', 
    'emAcompanhamentoNovos', 
    'interior/metropolitana'])

b2022_meses
```

### 3.7. 2022: percentual do aumento de casos entre semestres

```python
casos_2022_1 = b2022_1_ultimo_dia['casosAcumulado'].tolist()

casos_2022_2 = b2022_2_ultimo_dia['casosAcumulado'].tolist()

casos_1_2 = pd.DataFrame({
    'estados': estados,
    'número de casos 2022.1': casos_2022_1,
    'número de casos 2022.2': casos_2022_2})

percentual_aumento_casos = (casos_1_2['número de casos 2022.2']/casos_1_2['número de casos 2022.1'])*100

casos_1_2['percentual de aumento de casos'] = percentual_aumento_casos.round(2)
```

Tabela com o percentual de aumento do número de casos entre o final do primeiro e do segundo semestres de 2022

```python
casos_1_2
```

### 3.8. 2022: percentual do aumento de óbitos entre semestres

```python
obitos_2022_1 = b2022_1_ultimo_dia['obitosAcumulado'].tolist()

obitos_2022_2 = b2022_2_ultimo_dia['obitosAcumulado'].tolist()

obitos_1_2 = pd.DataFrame({
    'estados': estados,
    'número de casos 2022.1': obitos_2022_1,
    'número de casos 2022.2': obitos_2022_2})

percentual_aumento_obitos = (obitos_1_2['número de casos 2022.2']/obitos_1_2['número de casos 2022.1'])*100

obitos_1_2['percentual de aumento de óbitos'] = percentual_aumento_obitos.round(2)
```

Tabela com o percentual de aumento do número de óbitos entre o final do primeiro e do segundo semestres de 2022

```python
obitos_1_2
```

### 3.9. 2022: gráfico do número acumulado de casos

Em 2022, Ceará, Paraíba e Rio Grande do Norte foram os estados com maior aumento de número de casos, confirmando que o alto número total de casos no gráfico de linhas para Bahia, Pernambuco e Ceará se dá pelo alto número populacional.
