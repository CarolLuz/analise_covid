### 2.9. 2020: informações mensais (primeiro dia dos meses)

```python
jan = b2020_1.loc[b2020_1['data']=='2020-01-01']
jan = jan.loc[jan['codmun'].isna()].sort_values(by='estado')

fev = b2020_1.loc[b2020_1['data']=='2020-02-01']
fev = fev.loc[fev['codmun'].isna()].sort_values(by='estado')

mar = b2020_1.loc[b2020_1['data']=='2020-03-01']
mar = mar.loc[mar['codmun'].isna()].sort_values(by='estado')

abr = b2020_1.loc[b2020_1['data']=='2020-04-01']
abr = abr.loc[abr['codmun'].isna()].sort_values(by='estado')

mai = b2020_1.loc[b2020_1['data']=='2020-05-01']
mai = mai.loc[mai['codmun'].isna()].sort_values(by='estado')

jun = b2020_1.loc[b2020_1['data']=='2020-06-01']
jun = jun.loc[jun['codmun'].isna()].sort_values(by='estado')

jul = b2020_1.loc[b2020_1['data']=='2020-07-01']
jul = jul.loc[jul['codmun'].isna()].sort_values(by='estado')

ago = b2020_2.loc[b2020_2['data']=='2020-08-01']
ago = ago.loc[ago['codmun'].isna()].sort_values(by='estado')

setb = b2020_2.loc[b2020_2['data']=='2020-09-01']
setb = setb.loc[setb['codmun'].isna()].sort_values(by='estado')

out = b2020_2.loc[b2020_2['data']=='2020-10-01']
out = out.loc[out['codmun'].isna()].sort_values(by='estado')

nov = b2020_2.loc[b2020_2['data']=='2020-11-01']
nov = nov.loc[nov['codmun'].isna()].sort_values(by='estado')

dez = b2020_2.loc[b2020_2['data']=='2020-12-01']
dez = dez.loc[dez['codmun'].isna()].sort_values(by='estado')

b2020_meses = pd.concat([
    jan, 
    fev, 
    mar, 
    abr, 
    mai, 
    jun, 
    jul, 
    ago, 
    setb, 
    out, 
    nov, 
    dez], axis=0)

b2020_meses = b2020_meses.sort_values('data')

b2020_meses
```

### 2.10. 2020: informações mensais - removendo colunas não informativas

### 3.7. 2020: percentual do aumento de casos entre semestres

```python
casos_2020_1 = b2020_1_ultimo_dia['casosAcumulado'].tolist()

casos_2020_2 = b2020_2_ultimo_dia['casosAcumulado'].tolist()

casos_1_2 = pd.DataFrame({
    'estados': estados,
    'número de casos 2020.1': casos_2020_1,
    'número de casos 2020.2': casos_2020_2})

percentual_aumento_casos = (casos_1_2['número de casos 2020.2']/casos_1_2['número de casos 2020.1'])*100

casos_1_2['percentual de aumento de casos'] = percentual_aumento_casos.round(2)
```

Tabela com o percentual de aumento do número de casos entre o final do primeiro e do segundo semestres de 2020

```python
casos_1_2
```

### 3.8. 2020: percentual do aumento de óbitos entre semestres

```python
obitos_2020_1 = b2020_1_ultimo_dia['obitosAcumulado'].tolist()

obitos_2020_2 = b2020_2_ultimo_dia['obitosAcumulado'].tolist()

obitos_1_2 = pd.DataFrame({
    'estados': estados,
    'número de casos 2020.1': obitos_2020_1,
    'número de casos 2020.2': obitos_2020_2})

percentual_aumento_obitos = (obitos_1_2['número de casos 2020.2']/obitos_1_2['número de casos 2020.1'])*100

obitos_1_2['percentual de aumento de óbitos'] = percentual_aumento_obitos.round(2)
```

Tabela com o percentual de aumento do número de óbitos entre o final do primeiro e do segundo semestres de 2020

```python
obitos_1_2
```
